```python
import numpy as nd
import pandas as pd
```

```python
data = pd.read_csv("LA_TRANSITION_ECOLOGIQUE.csv")
```

```python
data = data.assign(createdAt=pd.to_datetime(data.createdAt),
                   publishedAt=pd.to_datetime(data.publishedAt),
                   updatedAt=pd.to_datetime(data.updatedAt))
```

```python
questions = data[data.columns[11:27]]
raw_text = ""
for i in range(len(questions.columns)):
    print(i)
    for answer in questions[questions.columns[i]].dropna():
        try:
            if len(answer) <= 10:
                continue
            raw_text += answer + " &&& "
        except:
            continue
    break
```
